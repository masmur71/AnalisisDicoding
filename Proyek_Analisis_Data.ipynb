## Exploratory Data Analysis (EDA)

###Pertanyaan 1

```python
# Buat Kolom baru yang mengkategorikan hari libur, hari kerja dan akhir minggu
day_df['day_type'] = day_df.apply(
    lambda row: 'Holiday' if row['holiday'] == 1 else ('Weekend' if row['workingday'] == 0 else 'Working Day'), axis=1
)

# Mengumpulkan data untuk menemukan jumlah rata-rata persewaan berdasarkan jenis hari
avg_rentals_by_day_type = day_df.groupby('day_type')['cnt'].mean().reset_index()
avg_rentals_by_day_type
```

###Pertanyaan 2

```python
# Membuat kolom baru untuk mengkategorikan hari sebagai 'Hari Kerja', 'Akhir Pekan', atau 'Libur'
hour_df['day_type'] = hour_df.apply(
    lambda row: 'Holiday' if row['holiday'] == 1 else ('Weekend' if row['workingday'] == 0 else 'Working Day'), axis=1
)

# Mengumpulkan data untuk membandingkan pengguna biasa dan pengguna terdaftar berdasarkan jenis hari (hari libur vs hari kerja)
avg_casual_registered_by_day_type = hour_df.groupby(['day_type'])[['casual', 'registered']].mean().reset_index()
avg_casual_registered_by_day_type
```

**Insight:**
- Pada pertanyaan 1, Ada perubahan pola penyewaan sepeda antara hari kerja, akhir pekan, dan hari libur. Penyewaan sepeda lebih banyak terjadi pada hari kerja dibandingkan dengan akhir pekan dan hari libur, kemungkinan karena banyak pengguna yang menggunakan sepeda sebagai alat transportasi untuk bekerja. Pada hari libur, penyewaan sepeda cenderung menurun, mungkin karena kegiatan atau rencana perjalanan masyarakat lebih sedikit dibandingkan pada hari kerja dan akhir pekan


- Pada pertanyaan 2, Pengguna casual cenderung lebih banyak menyewa sepeda pada akhir pekan dan hari libur dibandingkan dengan hari kerja. Pengguna registered lebih banyak menyewa sepeda pada hari kerja, yang mungkin menunjukkan bahwa mereka menggunakan sepeda sebagai transportasi utama untuk bekerja.

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=avg_rentals_by_day_type, x='day_type', y='cnt', palette='Blues_d')
plt.title('Average Bike Rentals by Day Type (Working Day, Weekend, Holiday)', fontsize=14)
plt.ylabel('Average Rentals')
plt.xlabel('Day Type')
plt.tight_layout()
plt.show()
```

### Pertanyaan 2:

```python
# Plotting the comparison
plt.figure(figsize=(10, 6))
avg_casual_registered_by_day_type.set_index('day_type')[['casual', 'registered']].plot(kind='bar', figsize=(10, 6), color=['skyblue', 'lightcoral'])

plt.title('Average Bike Rentals: Casual vs Registered Users (Working Day, Weekend, Holiday)', fontsize=14)
plt.ylabel('Average Rentals')
plt.xlabel('Day Type')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

**Insight:**
- Pertanyaan 1, Ada perbedaan pola penyewaan sepeda yang signifikan antara hari kerja, akhir pekan, dan hari libur. Sepeda lebih sering disewa pada hari kerja, menandakan bahwa sistem bike-sharing lebih banyak digunakan untuk transportasi sehari-hari daripada rekreasi. Sebaliknya, pada akhir pekan dan hari libur, penggunaan sepeda menurun, terutama pada hari libur, mungkin karena orang-orang tidak terlalu bergantung pada transportasi untuk bepergian di hari tersebut.
